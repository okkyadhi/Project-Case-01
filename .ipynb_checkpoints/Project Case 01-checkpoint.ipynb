Import pandas and numpy

```python
import pandas as pd
import numpy as np
```

Import germand credit score data 

```python
german = pd.read_csv('dataset/german_credit_data_with_risk.csv')
df = pd.DataFrame(german)
df
```

Check  null values

```python
# df.isnull().sum()
null_column = df.isnull().sum()
null_column[null_column>0]
```

normalize data

```python
df['Checking account'] = df['Checking account'].map({'little':1,'moderate':2,'rich':3,np.nan:0})
df
```

```python
df['Saving accounts'] = df['Saving accounts'].map({'little':1,'moderate':2,'quite rich':3,'rich':4,np.nan:0})
df
```

### 2. What is the minimum, maximum, and mean of the customer's age?

### 9. How many customers are classified as good risk and bad risk?

### 10. Is the average of credit amount for good risk is higher than bad risk?

## Part B


### 1. Who has the longest duration of installment?


```python
df[df['Duration']==df['Duration'].max()]
```

### 2. What is the most popular purpose?

```python
df.groupby('Purpose')['Purpose'].count().sort_values().tail(1)
```

### 3. Pearson Correlation between Risk and Saving Accounts

```python
df['Risk'] = df['Risk'].map({'good':1,'bad':0})
df
```

```python
df[['Risk','Saving accounts']].corr(method ='pearson')
```

### 4. How many unskilled and non-resident customer buy a car?


```python
pd.DataFrame({'Car':df[(df.Job == 0) & (df.Purpose == 'car') ][['Purpose']].count()})
```

### 5. How many rich customer does not have house?

```python
print('Rich customer who does not have own house : ',len(df[(df['Saving accounts']==4 )&(df.Housing != 'own')]))
```

### 6. Is rich customer always have a good risk?

```python
percent=df.groupby(['Saving accounts','Risk'])['Saving accounts'].size()
display(pd.DataFrame({'percent': percent.groupby(level=0).apply(lambda x: 100 * x / float(x.sum()))}))
print('Rich customer has the best risk score. There are 87% rich customer who have good risk')
```

### 7. Is Male has better risk score?

```python
percent = df.groupby(['Sex','Risk'])['Risk'].size()
display(pd.DataFrame(percent.groupby(level=0).apply(lambda x: x * 100/ x.sum())))
print('Male has better risk score than Female. 72% of Male has good risk')
```

### 8. How much credit amount  in bad risk customer?

```python
print('Total credit amount in bad risk customer: ' ,df[df.Risk==0]['Credit amount'].sum())
```

### 9. What average duration installment for Male Customer who had bad risk , owned house and duration installment is more than 12?

```python
df[(df.Risk == 0) & (df.Housing == 'own') & (df.Sex == 'male') & (df.Duration>12)]['Duration'].mean()
```

### 10. How is the correlation between credit amount and duration installment?

```python
display(df[['Duration','Credit amount']].corr())
print('Duration and credit amount has highly positive correlation means longest duration, higher credit amount')
```
